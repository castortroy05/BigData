```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(color_codes=True)

%matplotlib inline
```

```python
df = pd.read_csv("cardata.csv")
print(df.shape)
```

```python
df.head(5)
```

```python
print(df.dtypes)
```

```python
duplicate_rows_df = df[df.duplicated()]
print(duplicate_rows_df)
```

```python
df = df.drop_duplicates()
df.count()
```

```python
df = df.dropna()
print(df.count())
print(df.shape)
```

```python
print(df.describe())
```

```python
df.Make.value_counts().nlargest(40).plot(kind='bar', figsize=(10,5))
plt.title("Number of cars by make")
plt.ylabel('Number of cars')
plt.xlabel('Make')
```

```python
plt.hist(df['Engine HP'], bins = 20)
plt.title("Histogram of Engine HP")
plt.ylabel('Number of Engine HP')
plt.xlabel('Value of Engine HP')
```

```python
plt.figure(figsize = (20, 10))
corl = df.corr()
sns.heatmap(corl, cmap = "BrBG", annot = True)
print(corl)
```

```python
fig, ax = plt.subplots(figsize = (10, 6))
ax.scatter(df['Engine HP'], df['Price'])
ax.set_xlabel('Engine HP')
ax.set_ylabel('Price')
plt.show()
```
